```python
canada_geo = r'canada_provinces.json'
# create a plain world map
canada_map = folium.Map(location=[56.1304, -106.3468], zoom_start=2, tiles='Mapbox Bright')
print(canada_geo)
```

```python
# generate choropleth map using the total immigration of each country to Canada from 1980 to 2013
canada_map.choropleth(
    geo_data=canada_geo,
    data=canada,
    columns=['abbreviation', 'Population'],
    key_on="features.properties.District",
    fill_color='YlOrRd',
    fill_opacity=0.7, 
    line_opacity=0.2,
    legend_name='Population'
    )

# display map
canada_map
```

```python
results = requests.get(URL).json()
```

```python
venues = results['response']['venues']

BC_data = json_normalize(venues)
BC_data.head()
```

```python
# keep only columns that include venue name, and anything that is associated with location
filtered_columns = ['name', 'categories'] + [col for col in BC_data.columns if col.startswith('location.')] + ['id']
BC_data_filtered = BC_data.loc[:, filtered_columns]

# function that extracts the category of the venue
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']

# filter the category for each row
BC_data_filtered['categories'] = BC_data_filtered.apply(get_category_type, axis=1)

# clean column names by keeping only last term
BC_data_filtered.columns = [column.split('.')[-1] for column in BC_data_filtered.columns]

BC_data_filtered = BC_data_filtered[BC_data_filtered.distance != 10924]
BC_data_filtered.head()
```

```python
BC_data_filtered.categories.unique()
```

```python
venues_map = folium.Map(location=[latitude, longitude], zoom_start=13) # generate map centred around

# add a red circle marker to represent the Conrad Hotel
folium.features.CircleMarker(
    [latitude, longitude],
    radius=5,
    color='red',
    popup='University of British Columbia',
    fill = True,
    fill_color = 'red',
    fill_opacity = 0.6
).add_to(venues_map)

# add the Italian restaurants as blue circle markers
for lat, lng, label in zip(BC_data_filtered.lat, BC_data_filtered.lng, BC_data_filtered.categories):
    folium.features.CircleMarker(
        [lat, lng],
        radius=3,
        color='blue',
        popup=label,
        fill = True,
        fill_color='blue',
        fill_opacity=0.6
        ).add_to(venues_map)

# display map
venues_map
```
